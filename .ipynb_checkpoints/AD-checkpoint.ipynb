```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Lecture des fichiers CSV
catalogue = pd.read_csv("./Datasets/SD/Catalogue_Complet_Clean.csv")
clients = pd.read_csv("./Datasets/SD/clients - clean.csv")
marketing = pd.read_csv("./Datasets/SD/Marketing_clean.csv")

# Configuration des graphiques
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# === 1. Histogramme : Distribution du prix des vhicules ===
sns.histplot(catalogue['prix'], bins=20, kde=True, color="skyblue")
plt.title("Distribution du Prix des Vhicules")
plt.xlabel("Prix ()")
plt.ylabel("Nombre de vhicules")
plt.show()

# === 2. Histogramme : Puissance des vhicules ===
sns.histplot(catalogue['puissance'], bins=20, kde=True, color="lightgreen")
plt.title("Distribution de la Puissance des Vhicules")
plt.xlabel("Chevaux DIN")
plt.ylabel("Nombre de vhicules")
plt.show()

# === 3. Countplot : Rpartition par marque ===
plt.figure(figsize=(12, 6))
sns.countplot(data=catalogue, y='marque', order=catalogue['marque'].value_counts().index, palette="coolwarm")
plt.title("Nombre de vhicules par Marque")
plt.xlabel("Nombre")
plt.ylabel("Marque")
plt.show()

# === 4. Boxplot : Prix par Longueur du vhicule ===
sns.boxplot(data=catalogue, x='longueur', y='prix', palette="Set2")
plt.title("Prix des vhicules selon leur Longueur")
plt.xlabel("Longueur")
plt.ylabel("Prix ()")
plt.show()

# === 5. Countplot : Couleur des vhicules ===
sns.countplot(data=catalogue, x='couleur')
plt.title("Rpartition des vhicules par Couleur")
plt.xlabel("Couleur")
plt.ylabel("Nombre")
plt.show()

# === 6. Pie chart : Sexe des clients ===
clients['sexe'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'])
plt.title("Rpartition des clients selon le sexe")
plt.ylabel("")
plt.show()

# === 7. Histogramme : ge des clients ===
sns.histplot(clients['age'], bins=15, kde=True, color="orange")
plt.title("Distribution de l'ge des clients")
plt.xlabel("ge")
plt.ylabel("Nombre de clients")
plt.show()

# === 8. Countplot : Situation familiale des clients ===
sns.countplot(data=clients, x='situationFamiliale', palette="Set3")
plt.title("Rpartition par Situation Familiale")
plt.xlabel("Situation")
plt.ylabel("Nombre de clients")
plt.xticks(rotation=45)
plt.show()

# === 9. Nuage de points : Taux d'endettement vs ge ===
sns.scatterplot(data=clients, x='age', y='taux', hue='sexe')
plt.title("Taux dendettement en fonction de lge")
plt.xlabel("ge")
plt.ylabel("Taux ()")
plt.show()
```
