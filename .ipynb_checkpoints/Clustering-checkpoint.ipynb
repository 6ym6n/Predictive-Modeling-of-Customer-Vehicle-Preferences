```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./catalogue_unique_voitures clean.csv")
```

```python
print("=== ANALYSE DU DATASET ===")
print(f"Nombre de voitures: {len(df)}")
print(f"Colonnes: {list(df.columns)}")
print(df.describe())
```

```python
# 2. Sélection des features numériques
features = ['puissance', 'longueur']
X = df[features].copy()
```

```python
# 3. Normalisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 4. Méthode du coude pour déterminer k optimal
inertias = []
ks = range(1, 11)
for k in ks:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(ks, inertias, 'o-', linewidth=2)
plt.title("Méthode du coude pour K-Means")
plt.xlabel("Nombre de clusters K")
plt.ylabel("Inertie")
plt.xticks(ks)
plt.show()
```

```python
# Nombre de clusters fixé à 6
best_k = 4

print(f"\n=== CLUSTERING KMEANS AVEC k={best_k} ===")
kmeans = KMeans(n_clusters=best_k, random_state=9)
clusters = kmeans.fit_predict(X_scaled)
df['cluster'] = clusters
```

```python
# Analyse des clusters pour aide à la nomination
cluster_analysis = {}
for i in range(best_k):
    cluster_data = df[df['cluster'] == i]
    stats = {
        'puissance_moyenne': cluster_data['puissance'].mean(),
        'longueur_moyenne': cluster_data['longueur'].mean(),
        'nbPlaces_mode': cluster_data['nbPlaces'].mode().iloc[0],
        'nbPortes_mode': cluster_data['nbPortes'].mode().iloc[0],
        'marques': list(cluster_data['marque'].value_counts().index[:3])
    }
    cluster_analysis[i] = stats
    print(f"\n--- CLUSTER {i} ---")
    print(f"Voitures: {len(cluster_data)}")
    print(f"Puissance moyenne: {stats['puissance_moyenne']:.1f}")
    print(f"Longueur moyenne: {stats['longueur_moyenne']:.1f}")
    print(f"Places: {stats['nbPlaces_mode']}")
    print(f"Portes: {stats['nbPortes_mode']}")
    print(f"Marques principales: {', '.join(stats['marques'])}")
```

```python
# t-SNE
print("\n=== RÉDUCTION DE DIMENSION AVEC T-SNE ===")
tsne = TSNE(n_components=2, random_state=42, perplexity=6, n_iter=1000)
X_tsne = tsne.fit_transform(X_scaled)
df['tsne_x'] = X_tsne[:, 0]
df['tsne_y'] = X_tsne[:, 1]
```

```python
# Attribution de noms selon les observations ci-dessus (à ajuster selon ton jeu de données)
cluster_id_to_name = {
    0: "Familiales Puissantes",
    1: "Compactes Polyvalentes",
    2: "Petites Citadines",
    3: "Compactes Dynamiques"
}

df['categorie'] = df['cluster'].map(cluster_id_to_name)
```

```python
# Couleurs
colors = sns.color_palette("tab10", best_k)

# === PLOT 1 : Silhouette ===
plt.figure(figsize=(6, 4))
plt.bar([best_k], [silhouette_score(X_scaled, clusters)], color='skyblue')
plt.title(f"Score silhouette k = {best_k}")
plt.xlabel("k")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```

```python
 # === PLOT 2 : t-SNE ===
plt.figure(figsize=(8, 6))
for i in range(best_k):
    cluster_data = df[df['cluster'] == i]
    plt.scatter(cluster_data['tsne_x'], cluster_data['tsne_y'],
                label=cluster_id_to_name[i], alpha=0.7, s=60, c=[colors[i]])
plt.title("t-SNE avec K-Means")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# === PLOT 3 : Puissance vs Longueur ===
plt.figure(figsize=(8, 6))
for i in range(best_k):
    cluster_data = df[df['cluster'] == i]
    plt.scatter(cluster_data['puissance'], cluster_data['longueur'],
                label=cluster_id_to_name[i], alpha=0.7, s=60, c=[colors[i]])
plt.xlabel("Puissance")
plt.ylabel("Longueur")
plt.title("Puissance vs Longueur")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# === PLOT 4 : Nombre de voitures par catégorie ===
plt.figure(figsize=(7, 5))
df['categorie'].value_counts().plot(kind='bar', color=colors)
plt.title("Nombre de voitures par catégorie")
plt.xlabel("Catégorie")
plt.ylabel("Nombre")
plt.tight_layout()
plt.show()
```

```python
# Export
df_result = df[['marque', 'nom', 'puissance', 'longueur', 'nbPlaces', 'nbPortes', 
                'cluster', 'categorie']]
df_result.to_csv("clustering_KMeans_k6.csv", index=False)
print("✅ Exporté sous clustering_KMeans_k6.csv")
```
